Resize 800x800 pixels to 80x80 pixels image


```python
image1 = cv2.imread('img/checkboard.png',1)
new_dimensions = (80,80)
resized_image = cv2.resize(image1, new_dimensions, interpolation=cv2.INTER_AREA)
cv2.imwrite('img/resized_image.png',resized_image)
Image(filename='img/resized_image.png')
```

reading images with opencv

```python
image2 = cv2.imread('img/checkboard.png',-1)
print(image2)
```

```python
cv2.imwrite('img/image_gray.png',image2)
Image(filename='img/image_gray.png')
```

display attributes

```python
print('image size is: ', image2.shape)
print('image size is: ', image1.shape)
```

```python
print('image size is: ', image2.dtype)
print('image size is: ', image1.dtype)
```
